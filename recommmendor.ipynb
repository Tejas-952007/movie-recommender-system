```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
# Stem tags before vectorization (reduces vocab, better similarity)
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
new_df['tags'] = new_df['tags'].apply(lambda t: ' '.join(ps.stem(w) for w in t.split()))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
# Keep sparse to save memory; cosine_similarity supports sparse
vectors = cv.fit_transform(new_df['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Skip full N×N matrix (saves memory/time); similarity computed on-demand in recommend()
```

```python
# Example: top-5 similar to first movie (on-demand, no full matrix)
sorted(list(enumerate(cosine_similarity(vectors[0], vectors).flatten())), reverse=True, key=lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    # On-demand similarity only (sparse-friendly); no full N×N matrix
    sim = cosine_similarity(vectors[movie_index], vectors).flatten()
    top = sorted(enumerate(sim), key=lambda x: x[1], reverse=True)[1:6]
    for i, _ in top:
        print(new_df.iloc[i].title)
```

```python
recommend('Avatar')
```
